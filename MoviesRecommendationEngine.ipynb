```python
import pandas as pd 
import numpy as np
from sklearn.neighbors import NearestNeighbors

import warnings
warnings.filterwarnings("ignore")
```

```python
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")

ratings.head()
```

```python
ratings = ratings.drop("timestamp",axis=1)
```

```python
df = pd.merge(movies,ratings,how="inner").drop("genres",axis=1)
df
```

```python
df = pd.pivot_table(data=df, columns="userId", index=["movieId","title"],values=["rating"])
df
```

```python
df = df.reset_index().set_index("movieId")
df = df.fillna(0)
df
```

## Model Implemetation

```python
def find_movies(df,movie_name,k):
    
    model = NearestNeighbors(n_neighbors=k+1, algorithm="brute")
    model.fit(df.drop("title",axis=1))

    movie_id = df[df["title"] == movie_name].index[0]
    
    k_neighbors = model.kneighbors( df[df.index == movie_id].drop("title",axis=1), return_distance=False )[0].tolist()

    movies_titles = []

    for idx in k_neighbors:
        if idx != movie_id:
            title = df.iloc[idx]["title"].values.tolist()[0]
            if title != movie_name:
                movies_titles.append(title)
    
    return movies_titles
```

## Find recommendations the movies you like

#### specify the movies name (movie name should be one of those in the data frame).
#### specify the number of movies you'd like the engine to recommend.

```python
# movie name
movie_name = 'Shawshank Redemption, The (1994)'

# number of movies you want the engine to recommend 
k = 10
```

```python
movies_to_watch = find_movies(df,movie_name,k)
print("movies recommended to you : ")
for i in range(k):
    print(i+1,movies_to_watch[i])
```
